```python
from pandas import read_csv

df = read_csv('Time_Series_For_Clustering_El_Paso_Final.csv', index_col=False)
```

```python
import re
import pandas as pd

non_decimal = re.compile(r'[^\d.]+')

list_of_training_df = []

for i in range(1, 524):
    
    list_of_row_components = []
    
    current_row = df.iloc[i]
    current_row = current_row.iloc[1:]
    
    for j, row in current_row.iteritems():
        
        current_row_components = row.split(', ')
        current_row_components_replaced = []

        for idx, component in enumerate(current_row_components):

            result = non_decimal.sub('', current_row_components[idx])
            current_row_components_replaced.append(float(result))

        list_of_row_components.append(current_row_components_replaced)

    var1 = []
    var2 = []
    var3 = []
    var4 = []
    var5 = []
    var6 = []
    varout = []

    for element in list_of_row_components:

        var1.append(element[0])
        var2.append(element[2])
        var3.append(element[3])
        var4.append(element[4])
        var5.append(element[5])
        var6.append(element[6])
        varout.append(element[1])

    dict_temp = {'var1': var1, 'var2': var2, 'var3': var3, 'var4': var4, 'var5': var5, 'var6': var6, 'varout': varout}

    df_temp = pd.DataFrame(dict_temp)

    list_of_training_df.append(df_temp)
```

# The cell below utilizes VAR on each individual time series for each separate bridge and appends their rmse results into individual lists

```python
def Average(lst):
    return sum(lst) / len(lst)
```

```python
from statsmodels.tsa.vector_ar.var_model import VAR
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from math import sqrt
from sklearn.metrics import mean_squared_error

rmse_VAR_dict = dict.fromkeys(['var1', 'var2', 'var3', 'var4', 'var5', 'var6', 'varout'])

for item, idx in rmse_VAR_dict.items():

        rmse_VAR_dict[item] = []
        
rmse_ES_dict = dict.fromkeys(['var1', 'var2', 'var3', 'var4', 'var5', 'var6', 'varout'])

for item, idx in rmse_ES_dict.items():

    rmse_ES_dict[item] = []
        
for training_example in training_examples:
    
    # Get data for var
    train = training_example.iloc[0:21].values.tolist()
    valid = training_example.iloc[21:]
    
    # link to help with varmax and double exponential smoothing
    
    # var model
    model_VAR = VAR(endog=train)
    model_VAR_fit = model_VAR.fit(trend='nc')

    # var prediction
    prediction_VAR = model_VAR_fit.forecast(model_VAR_fit.y, steps=len(valid))
    
    # double exponential smoothing model (need separate one for each variable)
    model_ES_var1 = ExponentialSmoothing(training_example.iloc[0:]['var1'].values.tolist(), trend='add', seasonal=None, damped=True)
    model_ES_var2 = ExponentialSmoothing(training_example.iloc[0:21]['var2'].values.tolist(), trend='add', seasonal=None, damped=True)
    model_ES_var3 = ExponentialSmoothing(training_example.iloc[0:21]['var3'].values.tolist(), trend='add', seasonal=None, damped=True)
    model_ES_var4 = ExponentialSmoothing(training_example.iloc[0:21]['var4'].values.tolist(), trend='add', seasonal=None, damped=True)
    model_ES_var5 = ExponentialSmoothing(training_example.iloc[0:21]['var5'].values.tolist(), trend='add', seasonal=None, damped=True)
    model_ES_var6 = ExponentialSmoothing(training_example.iloc[0:21]['var6'].values.tolist(), trend='add', seasonal=None, damped=True)
    model_ES_varout = ExponentialSmoothing(training_example.iloc[0:21]['varout'].values.tolist(), trend='add', seasonal=None, damped=True)
   
    model_ES_var1_fit = model_ES_var1.fit()
    model_ES_var2_fit = model_ES_var2.fit()
    model_ES_var3_fit = model_ES_var3.fit()
    model_ES_var4_fit = model_ES_var4.fit()
    model_ES_var5_fit = model_ES_var5.fit()
    model_ES_var6_fit = model_ES_var6.fit()
    model_ES_varout_fit = model_ES_varout.fit()

    # double exponential smoothing prediction
    prediction_ES_var1 = model_ES_var1_fit.forecast(1)
    prediction_ES_var2 = model_ES_var2_fit.forecast(1)
    prediction_ES_var3 = model_ES_var3_fit.forecast(1)
    prediction_ES_var4 = model_ES_var4_fit.forecast(1)
    prediction_ES_var5 = model_ES_var5_fit.forecast(1)
    prediction_ES_var6 = model_ES_var6_fit.forecast(1)
    prediction_ES_varout = model_ES_varout_fit.forecast(1)
    
    # append predictions for VAR into list
    prediction_VAR_list = []
    
    for p in prediction_VAR:
        prediction_VAR_list.append(p.tolist())
        
    # append predictions for double exponential smoothing into list
    prediction_ES_var1_list = []
    prediction_ES_var2_list = []
    prediction_ES_var3_list = []
    prediction_ES_var4_list = []
    prediction_ES_var5_list = []
    prediction_ES_var6_list = []
    prediction_ES_varout_list = []
        
    # create dict for pred_VAR
    pred_VAR_dict = dict.fromkeys(['var1', 'var2', 'var3', 'var4', 'var5', 'var6', 'varout'])

    for item, idx in pred_VAR_dict.items():

        pred_VAR_dict[item] = []
        
    # create dict for pred_ES
    pred_ES_dict = dict.fromkeys(['var1', 'var2', 'var3', 'var4', 'var5', 'var6', 'varout'])
    
    for item, idx in pred_ES_dict.items():
        
        pred_ES_dict[item] = []

    # append prediction values for each variable into pred_VAR_dict
    for pred in prediction_VAR_list:
        
        pred_VAR_dict['var1'].append(pred[0])
        pred_VAR_dict['var2'].append(pred[1])
        pred_VAR_dict['var3'].append(pred[2])
        pred_VAR_dict['var4'].append(pred[3])
        pred_VAR_dict['var5'].append(pred[4])
        pred_VAR_dict['var6'].append(pred[5])
        pred_VAR_dict['varout'].append(pred[6])
        
    # append prediction values for each variable into pred_ES_dict
    pred_ES_dict['var1'].append(prediction_ES_var1.tolist())
    pred_ES_dict['var2'].append(prediction_ES_var2.tolist())
    pred_ES_dict['var3'].append(prediction_ES_var3.tolist())
    pred_ES_dict['var4'].append(prediction_ES_var4.tolist())
    pred_ES_dict['var5'].append(prediction_ES_var5.tolist())
    pred_ES_dict['var6'].append(prediction_ES_var6.tolist())
    pred_ES_dict['varout'].append(prediction_ES_varout.tolist())

    cols = training_examples[8].columns

    #check rmse
    for i in cols:
        rmse_VAR_dict[i].append(sqrt(mean_squared_error(pred_VAR_dict[i], valid[i])))
        rmse_ES_dict[i].append(sqrt(mean_squared_error(pred_ES_dict[i], [valid.iloc[-1][i]])))
#         print('rmse value for', i, 'is : ', sqrt(mean_squared_error(pred_dict[i], valid[i])))
            
```

```python
from statsmodels.tsa.vector_ar.var_model import VAR
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from math import sqrt
from sklearn.metrics import mean_squared_error

rmse_VAR_dict = dict.fromkeys(['var1', 'var2', 'var3', 'var4', 'var5', 'var6', 'var7', 'var8', 'var9', 'var10', 'var11', 'var12', 'varout'])

for item, idx in rmse_VAR_dict.items():

        rmse_VAR_dict[item] = []
        
rmse_ES_dict = dict.fromkeys(['var1', 'var2', 'var3', 'var4', 'var5', 'var6', 'var7', 'var8', 'var9', 'var10', 'var11', 'var12', 'varout'])

for item, idx in rmse_ES_dict.items():

    rmse_ES_dict[item] = []
        
for training_example in training_examples:
    
    # Get data for var
    train = training_example.iloc[0:21].values.tolist()
    valid = training_example.iloc[21:]
    
    # link to help with varmax and double exponential smoothing
    
    # var model
    model_VAR = VAR(endog=train)
    model_VAR_fit = model_VAR.fit(trend='nc')

    # var prediction
    prediction_VAR = model_VAR_fit.forecast(model_VAR_fit.y, steps=len(valid))
    
    # double exponential smoothing model (need separate one for each variable)
    model_ES_var1 = ExponentialSmoothing(training_example.iloc[0:]['var1'].values.tolist(), trend='add', seasonal=None, damped=True)
    model_ES_var2 = ExponentialSmoothing(training_example.iloc[0:21]['var2'].values.tolist(), trend='add', seasonal=None, damped=True)
    model_ES_var3 = ExponentialSmoothing(training_example.iloc[0:21]['var3'].values.tolist(), trend='add', seasonal=None, damped=True)
    model_ES_var4 = ExponentialSmoothing(training_example.iloc[0:21]['var4'].values.tolist(), trend='add', seasonal=None, damped=True)
    model_ES_var5 = ExponentialSmoothing(training_example.iloc[0:21]['var5'].values.tolist(), trend='add', seasonal=None, damped=True)
    model_ES_var6 = ExponentialSmoothing(training_example.iloc[0:21]['var6'].values.tolist(), trend='add', seasonal=None, damped=True)
    model_ES_var7 = ExponentialSmoothing(training_example.iloc[0:21]['var7'].values.tolist(), trend='add', seasonal=None, damped=True)
    model_ES_var8 = ExponentialSmoothing(training_example.iloc[0:21]['var8'].values.tolist(), trend='add', seasonal=None, damped=True)
    model_ES_var9 = ExponentialSmoothing(training_example.iloc[0:21]['var9'].values.tolist(), trend='add', seasonal=None, damped=True)
    model_ES_var10 = ExponentialSmoothing(training_example.iloc[0:21]['var10'].values.tolist(), trend='add', seasonal=None, damped=True)
    model_ES_var11 = ExponentialSmoothing(training_example.iloc[0:21]['var11'].values.tolist(), trend='add', seasonal=None, damped=True)
    model_ES_var12 = ExponentialSmoothing(training_example.iloc[0:21]['var12'].values.tolist(), trend='add', seasonal=None, damped=True)
    
    model_ES_varout = ExponentialSmoothing(training_example.iloc[0:21]['varout'].values.tolist(), trend='add', seasonal=None, damped=True)
   
    model_ES_var1_fit = model_ES_var1.fit()
    model_ES_var2_fit = model_ES_var2.fit()
    model_ES_var3_fit = model_ES_var3.fit()
    model_ES_var4_fit = model_ES_var4.fit()
    model_ES_var5_fit = model_ES_var5.fit()
    model_ES_var6_fit = model_ES_var6.fit()
    model_ES_var7_fit = model_ES_var7.fit()
    model_ES_var8_fit = model_ES_var8.fit()
    model_ES_var9_fit = model_ES_var9.fit()
    model_ES_var10_fit = model_ES_var10.fit()
    model_ES_var11_fit = model_ES_var11.fit()
    model_ES_var12_fit = model_ES_var12.fit()
                               
    model_ES_varout_fit = model_ES_varout.fit()

    # double exponential smoothing prediction
    prediction_ES_var1 = model_ES_var1_fit.forecast(1)
    prediction_ES_var2 = model_ES_var2_fit.forecast(1)
    prediction_ES_var3 = model_ES_var3_fit.forecast(1)
    prediction_ES_var4 = model_ES_var4_fit.forecast(1)
    prediction_ES_var5 = model_ES_var5_fit.forecast(1)
    prediction_ES_var6 = model_ES_var6_fit.forecast(1)
    prediction_ES_var7 = model_ES_var7_fit.forecast(1)
    prediction_ES_var8 = model_ES_var8_fit.forecast(1)
    prediction_ES_var9 = model_ES_var9_fit.forecast(1)
    prediction_ES_var10 = model_ES_var10_fit.forecast(1)
    prediction_ES_var11 = model_ES_var11_fit.forecast(1)
    prediction_ES_var12 = model_ES_var12_fit.forecast(1)
    prediction_ES_varout = model_ES_varout_fit.forecast(1)
    
    # append predictions for VAR into list
    prediction_VAR_list = []
    
    for p in prediction_VAR:
        prediction_VAR_list.append(p.tolist())
        
    # append predictions for double exponential smoothing into list
    prediction_ES_var1_list = []
    prediction_ES_var2_list = []
    prediction_ES_var3_list = []
    prediction_ES_var4_list = []
    prediction_ES_var5_list = []
    prediction_ES_var6_list = []
    prediction_ES_var7_list = []
    prediction_ES_var8_list = []
    prediction_ES_var9_list = []
    prediction_ES_var10_list = []
    prediction_ES_var11_list = []
    prediction_ES_var12_list = []
    prediction_ES_varout_list = []
        
    # create dict for pred_VAR
    pred_VAR_dict = dict.fromkeys(['var1', 'var2', 'var3', 'var4', 'var5', 'var6', 'var7', 'var8', 'var9', 'var10', 'var11', 'var12', 'varout'])

    for item, idx in pred_VAR_dict.items():

        pred_VAR_dict[item] = []
        
    # create dict for pred_ES
    pred_ES_dict = dict.fromkeys(['var1', 'var2', 'var3', 'var4', 'var5', 'var6', 'var7', 'var8', 'var9', 'var10', 'var11', 'var12', 'varout'])
    
    for item, idx in pred_ES_dict.items():
        
        pred_ES_dict[item] = []

    # append prediction values for each variable into pred_VAR_dict
    for pred in prediction_VAR_list:
        
        pred_VAR_dict['var1'].append(pred[0])
        pred_VAR_dict['var2'].append(pred[1])
        pred_VAR_dict['var3'].append(pred[2])
        pred_VAR_dict['var4'].append(pred[3])
        pred_VAR_dict['var5'].append(pred[4])
        pred_VAR_dict['var6'].append(pred[5])
        pred_VAR_dict['var7'].append(pred[6])
        pred_VAR_dict['var8'].append(pred[7])
        pred_VAR_dict['var9'].append(pred[8])
        pred_VAR_dict['var10'].append(pred[9])
        pred_VAR_dict['var11'].append(pred[10])
        pred_VAR_dict['var12'].append(pred[11])
        pred_VAR_dict['varout'].append(pred[6])
        
    # append prediction values for each variable into pred_ES_dict
    pred_ES_dict['var1'].append(prediction_ES_var1.tolist())
    pred_ES_dict['var2'].append(prediction_ES_var2.tolist())
    pred_ES_dict['var3'].append(prediction_ES_var3.tolist())
    pred_ES_dict['var4'].append(prediction_ES_var4.tolist())
    pred_ES_dict['var5'].append(prediction_ES_var5.tolist())
    pred_ES_dict['var6'].append(prediction_ES_var6.tolist())
    pred_ES_dict['var7'].append(prediction_ES_var7.tolist())
    pred_ES_dict['var8'].append(prediction_ES_var8.tolist())
    pred_ES_dict['var9'].append(prediction_ES_var9.tolist())
    pred_ES_dict['var10'].append(prediction_ES_var10.tolist())
    pred_ES_dict['var11'].append(prediction_ES_var11.tolist())
    pred_ES_dict['var12'].append(prediction_ES_var12.tolist())
    pred_ES_dict['varout'].append(prediction_ES_varout.tolist())

    cols = training_examples[8].columns

    #check rmse
    for i in cols:
        rmse_VAR_dict[i].append(sqrt(mean_squared_error(pred_VAR_dict[i], valid[i])))
        rmse_ES_dict[i].append(sqrt(mean_squared_error(pred_ES_dict[i], [valid.iloc[-1][i]])))
#         print('rmse value for', i, 'is : ', sqrt(mean_squared_error(pred_dict[i], valid[i])))
            
```
